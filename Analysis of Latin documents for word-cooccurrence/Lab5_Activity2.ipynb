```r
library(ggplot2)
```

```r
x<-c(5,10,15,20,25,30,35,40)
y<-c(264,329,376,450,505,631,702,680)
```

```r
data<-data.frame(x,y)
```

```r
head(data)
```

```r
ggplot(data, aes(x,y)) + geom_line(colour = 'red')+ geom_point(colour='blue')+ xlab("Files") +
  ylab("Time(sec)") + ggtitle("Bigrams") + geom_text(aes(label=y),hjust=0, vjust=0)+scale_x_continuous(limits = c(0, 60))
```

```r
x1<-c(5,10,15,20,25,30,35,40)
y1<-c(255,330,400,501,520,650,710,760)
```

```r
data2<-data.frame(x1,y1)
```

```r
ggplot(data2, aes(x1,y1)) + geom_line(colour = 'red')+ geom_point(colour='blue')+ xlab("Files") +
  ylab("Time(sec)") + ggtitle("Trigrams") + geom_text(aes(label=y1),hjust=0, vjust=0)+scale_x_continuous(limits = c(0, 60))
```
